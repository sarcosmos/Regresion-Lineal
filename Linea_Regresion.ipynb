# *Descripción:*

Área: *salud*

En el ejercicio propuesto vamos a ver la relación entre el colesterol y las horas de ejercicio.



---






```python
# Importamos las bibliotecas necesarias
import numpy as np  # Para trabajar con arreglos numéricos y realizar cálculos matemáticos
from sklearn.model_selection import train_test_split  # Para dividir los datos en conjuntos de entrenamiento y prueba
from sklearn.linear_model import LinearRegression  # Para crear el modelo de regresión lineal
import matplotlib.pyplot as plt  # Para visualizar los datos y resultados

# Creamos los datos de horas de ejercicio y niveles de colesterol
exercise_hours = np.array([2, 5, 4, 1, 6, 3, 7, 8, 2, 5, 9, 6, 4, 7, 3, 8, 9, 1, 5, 6, 2, 3, 7, 8, 4, 5, 6, 7, 8])
cholesterol_level = np.array([210, 190, 200, 220, 180, 205, 170, 165, 215, 185, 160, 175, 200, 170, 210, 160, 155, 225, 195, 180, 220, 210, 175, 165, 205, 190,185,175,160])

# Redimensionamos el arreglo de horas de ejercicio para que sea una columna (necesario para sklearn)
x = exercise_hours.reshape(-1, 1)  # reshape(-1, 1) convierte el arreglo en una matriz columna
y = cholesterol_level  # Los niveles de colesterol permanecen sin cambios (arreglo 1D)

# Dividimos los datos en entrenamiento (80%) y prueba (20%)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# test_size=0.2 significa que el 20% de los datos se usarán para pruebas y el resto para entrenamiento
# random_state=42 asegura que la división sea reproducible

# Creamos una instancia del modelo de regresión lineal
model = LinearRegression()

# Entrenamos el modelo con los datos de entrenamiento (horas de ejercicio y niveles de colesterol)
model.fit(x_train, y_train)

# Realizamos predicciones usando los datos de prueba
y_pred = model.predict(x_test)

# Calculamos el puntaje R^2 (qué tan bien se ajusta el modelo a los datos de prueba)
r2 = model.score(x_test, y_test)

# Obtenemos el coeficiente (pendiente) y el intercepto del modelo
coefficient = model.coef_[0]  # El coeficiente indica cuánto cambia el colesterol por cada hora de ejercicio
intercept = model.intercept_  # El intercepto es el nivel de colesterol cuando no hay ejercicio

# Imprimimos los resultados del modelo
print("Linear Regression")
print("R2 score:", r2)  # R2 es una métrica que indica la calidad del ajuste del modelo (valor entre 0 y 1)
print("Coefficient:", coefficient)  # Imprime la pendiente de la línea de regresión
print("Intercept:", intercept)  # Imprime el valor de intersección con el eje Y

# Comparamos los valores actuales (reales) y predichos para el conjunto de prueba
print("Predicted vs Current")
for current, predicted in zip(y_test, y_pred):
    print(f"Current: {current:.2f}, Predicted: {predicted:.2f}")  # Compara valores actuales y predichos

# Graficamos los puntos reales y la línea de predicción
plt.scatter(x, y, color='blue', label='Actual')  # Los puntos azules representan los datos reales
plt.plot(x_test, y_pred, color='red', label='Predicted')  # La línea roja representa las predicciones
plt.xlabel('Exercise Hours')  # Etiqueta del eje X
plt.ylabel('Cholesterol Level')  # Etiqueta del eje Y
plt.title('Exercise Hours vs Cholesterol Level')  # Título del gráfico
plt.legend()  # Añade una leyenda para diferenciar entre datos reales y predicciones
plt.show()  # Muestra el gráfico
```

# *Conclusiones:*

Como podemos observar en la gráfica se puede ver la relación directa entre las horas de ejercicio y los niveles de colesterol, donde a mayor cantidad de horas de ejercicio se da una bajada de los niveles de colesterol en sangre.
